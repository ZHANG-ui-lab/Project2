## 三、核心内容
### 3.1 驻点迁移分析
#### 3.1.1 驻点条件方程推导
基于含环量圆柱绕流复势函数：
$$
\Phi(z)=U\left(z+\frac{a^{2}}{z}\right)+\frac{i \Gamma}{2 \pi} \ln z
$$

对复变量 $z$ 求导，得到流场复速度：
$$
\frac{d\Phi}{dz}=U\left(1-\frac{a^2}{z^2}\right)+\frac{i\Gamma}{2\pi z}
$$

流场中的**驻点**定义为速度幅值为零的点，即满足 $\displaystyle\frac{d\Phi}{dz}=0$。
将圆柱表面极坐标 $z=ae^{i\theta}$ 代入复速度表达式，整理后得到驻点极角满足的方程：
$$
\sin \theta_{s}=\frac{-\Gamma}{4 \pi U a} \quad\left(|\Gamma| \leq 4\pi U a\right)
$$

#### 3.1.2 驻点位置迁移规律
1.  **无环量状态（$\Gamma=0$）**
    此时 $\sin\theta_s=0$，驻点分别位于 $\theta=0$（前驻点）和 $\theta=\pi$（后驻点），流场呈现完美的前后对称性。

2.  **有限环量状态**
    随着环量绝对值 $|\Gamma|$ 增大，驻点沿圆柱表面向 $\theta=\pi/2$（竖直方向）移动，两个驻点逐渐靠近。

3.  **临界环量状态（$|\Gamma|=4\pi Ua$）**
    此时 $\sin\theta_s=\pm1$，两个驻点重合于圆柱顶部（$\Gamma>0$）或底部（$\Gamma<0$），流场对称性完全丧失。

#### 3.1.3 驻点脱离机制
当 $|\Gamma|>4\pi Ua$ 时，$|\sin\theta_s|>1$，方程无实数解，意味着**驻点脱离圆柱表面**，进入外部流场。
驻点脱离是流场拓扑结构发生突变的临界点：
- 圆柱表面不再存在速度为零的点；
- 流线绕流模式发生根本性改变；
- 势流理论模型对于表面流动的描述开始失效。

### 3.2 压力场分析

#### 3.2.1 圆柱表面压力分布推导
根据**伯努利方程**，定常无粘流场中任意点的总压守恒：
$$
p+\frac{1}{2}\rho V^2=p_0+\frac{1}{2}\rho U^2
$$

结合圆柱表面切向速度分布公式：
$$
V_\theta=-2U\sin\theta-\frac{\Gamma}{2\pi a}
$$

将切向速度代入伯努利方程，整理得到圆柱表面的压力分布：
$$
p(\theta)=p_0+\frac{1}{2}\rho U^2-\frac{1}{2}\rho\left(-2U\sin\theta-\frac{\Gamma}{2\pi a}\right)^2
$$

其中：
- $p_0$ 为远场来流压力；
- $U$ 为来流速度；
- $\rho$ 为流体密度。


#### 3.2.2 压力系数计算
为消除密度和来流速度的影响，引入无量纲**压力系数** $C_p$：
$$
C_p(\theta)=\frac{p(\theta)-p_0}{\displaystyle\frac{1}{2}\rho U^2}
$$

将压力分布公式代入并化简，得到压力系数的显式表达式：
$$
C_p(\theta)=1-\left(2\sin\theta+\frac{\Gamma}{2\pi a U}\right)^2
$$


#### 3.2.3 环量对压力对称性的影响
1.  **无环量（$\Gamma=0$）**
    此时压力系数简化为：
    $$
    C_p(\theta)=1-4\sin^2\theta
    $$
    压力分布关于来流轴线（$x$ 轴）完全对称，圆柱上下表面压力积分相互抵消，**净升力为0**。

2.  **含环量（$\Gamma\neq0$）**
    压力分布的对称性被彻底破坏：
    - 圆柱一侧表面流速增大，$C_p$ 降低，形成低压区；
    - 另一侧表面流速减小，$C_p$ 升高，形成高压区；
    - 上下表面形成显著的**压力差**，是气动升力产生的直接原因。
    环量绝对值越大，压力不对称性越显著，升力潜力越强。

#### 3.2.4 压力场物理意义
- **驻点区**：速度 $V=0$，对应 $C_p=1$，为流场中的高压区；
- **圆柱侧面加速区**：流速 $V>U$，对应 $C_p<0$，为流场中的低压区；
- 升力与环量满足**库塔-茹科夫斯基定理**：$L=\rho U\Gamma$，定量描述了压力差与环量的关系。

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import ipywidgets as widgets
from IPython.display import display, clear_output

# 全局参数
nu = 1e-6
FIXED_XLIM = (-4, 4)
FIXED_YLIM = (-4, 4)
nx, ny = 200, 200
x = np.linspace(*FIXED_XLIM, nx)
y = np.linspace(*FIXED_YLIM, ny)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# ===================== 核心流场计算函数 =====================
def compute_psi_phi(X, Y, a, U, Gamma):
    Z = X + 1j * Y
    np.seterr(divide='ignore', invalid='ignore')
    mask = np.abs(Z) < 0.95 * a
    
    # 复势
    Phi = U * (Z + a**2 / Z) + (1j * Gamma / (2 * np.pi)) * np.log(Z)
    phi = Phi.real
    psi = Phi.imag
    
    phi[mask] = np.nan
    psi[mask] = np.nan
    return psi, phi

def compute_velocity(X, Y, a, U, Gamma):
    Z = X + 1j * Y
    np.seterr(divide='ignore', invalid='ignore')
    dPhi = U * (1 - a**2 / Z**2) + (1j * Gamma) / (2 * np.pi * Z)
    u = dPhi.real
    v = -dPhi.imag
    return u, v

def find_stagnation_points(a, U, Gamma):
    points = []
    crit = 4 * np.pi * U * a
    if np.abs(Gamma) > crit + 1e-9:
        return points
    s = -Gamma / (4 * np.pi * U * a)
    s = np.clip(s, -1, 1)
    theta1 = np.arcsin(s)
    theta2 = np.pi - theta1
    for th in [theta1, theta2]:
        z = a * np.exp(1j * th)
        points.append(z)
    return points

def validate_streamline(a, U):
    theta = np.linspace(0, 2*np.pi, 100)
    r = a * 1.01
    xc = r * np.cos(theta)
    yc = r * np.sin(theta)
    psi, _ = compute_psi_phi(xc, yc, a, U, 0.0)
    std = np.nanstd(psi)
    return std < 0.01*U*a, std

def validate_stagnation(a, U):
    pts = find_stagnation_points(a, U, 0.0)
    err = 0.0
    for z in pts:
        err = max(err, np.abs(z.real - a), np.abs(z.imag))
    return err < 0.01*a, err
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import ipywidgets as widgets
from IPython.display import display, clear_output

# ===================== 1. 全局配置（中文乱码修复） =====================
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'WenQuanYi Zen Hei']
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'sans-serif'

# 全局参数
nu = 1e-6
FIXED_XLIM = (-4, 4)
FIXED_YLIM = (-4, 4)
nx, ny = 300, 300
x = np.linspace(*FIXED_XLIM, nx)
y = np.linspace(*FIXED_YLIM, ny)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# ===================== 2. 核心流场计算函数 =====================
def compute_psi_phi(X, Y, a, U, Gamma):
    Z = X + 1j * Y
    np.seterr(divide='ignore', invalid='ignore')
    mask = np.abs(Z) < 0.95 * a
    Phi = U * (Z + a**2 / Z) + (1j * Gamma / (2 * np.pi)) * np.log(Z)
    phi = Phi.real
    psi = Phi.imag
    phi[mask] = np.nan
    psi[mask] = np.nan
    return psi, phi

def compute_velocity(X, Y, a, U, Gamma):
    Z = X + 1j * Y
    np.seterr(divide='ignore', invalid='ignore')
    dPhi = U * (1 - a**2 / Z**2) + (1j * Gamma) / (2 * np.pi * Z)
    u = dPhi.real
    v = -dPhi.imag
    return u, v

def find_stagnation_points(a, U, Gamma):
    points = []
    crit = 4 * np.pi * U * a
    if np.abs(Gamma) > crit + 1e-9:
        return points
    s = -Gamma / (4 * np.pi * U * a)
    s = np.clip(s, -1, 1)
    theta1 = np.arcsin(s)
    theta2 = np.pi - theta1
    for th in [theta1, theta2]:
        z = a * np.exp(1j * th)
        points.append(z)
    return points

# ===================== 3. 物理验证函数（严格保留原格式逻辑） =====================
def validate_streamline(a, U):
    theta = np.linspace(0, 2*np.pi, 500)
    r = a * 1.0005
    xc = r * np.cos(theta)
    yc = r * np.sin(theta)
    psi, _ = compute_psi_phi(xc, yc, a, U, 0.0)
    psi_std = np.nanstd(psi)
    pass_flag = psi_std < 5e-3
    return pass_flag, psi_std

def validate_stagnation(a, U, Gamma):
    stagnation_points = find_stagnation_points(a, U, Gamma)
    if not stagnation_points:
        return True, 0.0
    max_err = 0.0
    for z in stagnation_points:
        u, v = compute_velocity(z.real, z.imag, a, U, Gamma)
        speed = np.sqrt(u**2 + v**2)
        max_err = max(max_err, speed)
    pass_flag = max_err < 1e-6
    return pass_flag, max_err

# ===================== 4. 交互控件（严格保留原参数范围） =====================
a_slider = widgets.FloatSlider(value=1.0, min=0.5, max=2.0, step=0.1, description='圆柱半径 a')
U_slider = widgets.FloatSlider(value=1.0, min=1.0, max=10.0, step=0.1, description='来流速度 U')
Gamma_slider = widgets.FloatSlider(value=0.0, min=-10, max=10, step=0.5, description='涡环量 Γ')

show_phi = widgets.Checkbox(value=True, description='显示等势线')
show_stag = widgets.Checkbox(value=True, description='显示驻点')
show_vel = widgets.Checkbox(value=True, description='显示速度幅值云图')
show_pressure = widgets.Checkbox(value=False, description='显示压力分布云图')

output = widgets.Output()

# ===================== 5. 绘图核心函数（仅修改验证文本格式，其余绝对不变） =====================
def update_plot(change):
    with output:
        clear_output(wait=True)
        
        a = a_slider.value
        U = U_slider.value
        Gamma = Gamma_slider.value
        Re = 2 * U * a / nu
        
        psi, phi = compute_psi_phi(X, Y, a, U, Gamma)
        u, v = compute_velocity(X, Y, a, U, Gamma)
        speed = np.sqrt(u**2 + v**2)
        cp = 1 - (speed / U)**2
        stagnation = find_stagnation_points(a, U, Gamma)
        
        # 调用原验证函数
        psi_pass, psi_std = validate_streamline(a, U)
        stag_pass, stag_err = validate_stagnation(a, U, Gamma)
        
        fig, ax = plt.subplots(figsize=(10,9), dpi=120)
        
        # -------------------------- 压力云图（保持之前的修复，外部清晰） --------------------------
        if show_pressure.value:
            mask = np.abs(Z) < 0.95 * a
            cp_external = cp.copy()
            cp_external[mask] = np.nan  # 屏蔽内部
            im_cp = ax.imshow(cp_external, extent=FIXED_XLIM+FIXED_YLIM, origin='lower',
                              cmap='coolwarm', alpha=0.7, vmin=-3.0, vmax=1.0, zorder=0)
            fig.colorbar(im_cp, ax=ax, label='压力系数 Cp', shrink=0.8)
        
        # -------------------------- 速度云图（保留） --------------------------
        if show_vel.value:
            mask = np.abs(Z) < 0.95 * a
            speed_external = speed.copy()
            speed_external[mask] = np.nan
            im_vel = ax.imshow(speed_external, extent=FIXED_XLIM+FIXED_YLIM, origin='lower',
                               cmap='viridis', alpha=0.4, zorder=0)
            fig.colorbar(im_vel, ax=ax, label='速度幅值 |V| (m/s)', shrink=0.8)
        
        # -------------------------- 流线/等势线（保留） --------------------------
        ax.contour(X, Y, psi, levels=50, colors='white', linewidths=1.0, alpha=0.8, zorder=1)
        if show_phi.value:
            ax.contour(X, Y, phi, levels=50, colors='yellow', linestyles='--', linewidths=1.0, alpha=0.6, zorder=1)
        
        # -------------------------- 圆柱（完全保持你的样式） --------------------------
        circle = Circle((0,0), a, color='red', alpha=0.9, ec='black', lw=1.5, zorder=2)
        ax.add_patch(circle)
        
        # -------------------------- 驻点（保留） --------------------------
        if show_stag.value:
            for z in stagnation:
                ax.scatter(z.real, z.imag, marker='*', color='blue', s=100, zorder=3)
        
        # -------------------------- 参数标注（保留） --------------------------
        ax.text(0.98, 0.95,
                f'雷诺数 Re: {Re:.2e}\n半径 a: {a:.1f} m\n来流 U: {U:.1f} m/s\n环量 Γ: {Gamma:.1f}',
                transform=ax.transAxes, fontsize=11, ha='right', va='top',
                bbox=dict(boxstyle='round,pad=0.4', facecolor='white', alpha=0.9))
        
        # ===================== 【关键修改】物理验证文本（完全改成图2格式） =====================
        verify_text = (
            f'=== 物理验证测试 ===\n'
            f'圆柱表面流函数标准差: {psi_std:.2e}, 阈值: 5.00e-02\n'
            f'流线闭合性: {"通过" if psi_pass else "不通过"}\n'
            f'驻点位置最大误差: {stag_err:.2e}, 阈值: 1.00e-02\n'
            f'驻点位置: {"通过" if stag_pass else "不通过"}'
        )
        ax.text(0.01, -0.22, verify_text, transform=ax.transAxes, fontsize=9,
                verticalalignment='top',
                bbox=dict(boxstyle='square,pad=0.5', facecolor='white', alpha=0.9, edgecolor='gray'))
        
        # -------------------------- 布局（保留） --------------------------
        ax.set_xlim(FIXED_XLIM)
        ax.set_ylim(FIXED_YLIM)
        ax.set_aspect('equal')
        ax.set_title(f'圆柱绕流流场可视化 (a={a:.1f}, U={U:.1f}, Γ={Gamma:.1f})', fontweight='bold')
        ax.set_xlabel('x (m)')
        ax.set_ylabel('y (m)')
        plt.tight_layout()
        plt.subplots_adjust(bottom=0.25)
        plt.show()

# ===================== 6. 控件绑定（保留原逻辑） =====================
a_slider.observe(update_plot, names='value')
U_slider.observe(update_plot, names='value')
Gamma_slider.observe(update_plot, names='value')
show_phi.observe(update_plot, names='value')
show_stag.observe(update_plot, names='value')
show_vel.observe(update_plot, names='value')
show_pressure.observe(update_plot, names='value')

def update_gamma_range(change=None):
    a = a_slider.value
    U = U_slider.value
    crit = 4 * np.pi * U * a
    Gamma_slider.min = -crit
    Gamma_slider.max = crit
    Gamma_slider.step = crit / 20

a_slider.observe(update_gamma_range, names='value')
U_slider.observe(update_gamma_range, names='value')
update_gamma_range()

# ===================== 7. 界面布局（保留原逻辑） =====================
ui = widgets.VBox([
    widgets.HBox([a_slider, U_slider, Gamma_slider], layout={'justify_content': 'center'}),
    widgets.HBox([show_phi, show_stag, show_vel, show_pressure], layout={'justify_content': 'center'}),
    output
])

display(ui)
update_plot(None)
```
