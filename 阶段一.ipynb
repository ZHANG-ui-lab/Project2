# 模块 2：带环量圆柱绕流复势构造
## 2 带环量圆柱绕流复势构造

### 2.1 基础流场：无环量圆柱绕流复势
均匀来流叠加偶极子，得到满足圆柱边界的无环量复势：

$$
\Phi_0(z)=U\left(z+\frac{a^2}{z}\right)
$$

该复势在 $r>a$ 内解析，流场对称，无环量、无升力。

### 2.2 点涡复势引入与物理意义
点涡是平面势流基本解，复势为多值复对数函数：

$$
\Phi_\Gamma(z)=\frac{i\Gamma}{2\pi}\ln z
$$

- $\Gamma$：环量，表征涡强度，是产生升力的核心参数；
- $\ln z$ 多值性：每绕原点一周，虚部跃变 $2\pi$，使速度势产生周期跃变；
- 不改变连续性与无旋性，仅打破流场对称性。

### 2.3 带环量总复势构建
线性叠加无环量复势与点涡复势，得到带环量圆柱绕流总复势：

$$
\Phi(z)=U\left(z+\frac{a^2}{z}\right)+\frac{i\Gamma}{2\pi}\ln z
$$

该式为本阶段核心数学模型，满足圆柱无穿透边界条件，是后续所有推导的基础。

### 2.4 极坐标展开与多值性分析
令 $z=re^{i\theta}$，将复势分离为速度势 $\varphi$（实部）与流函数 $\psi$（虚部）：

$$
\varphi=U\left(r+\frac{a^2}{r}\right)\cos\theta-\frac{\Gamma}{2\pi}\theta
$$

$$
\psi=U\left(r-\frac{a^2}{r}\right)\sin\theta+\frac{\Gamma}{2\pi}\ln r
$$

#### 多值性与周期跃变
复对数 $\ln z=\ln r+i(\theta+2k\pi)$ 具有多值性，当 $\theta$ 从 $0\to2\pi$（绕圆柱一周）：

$$
\Delta\varphi=-\Gamma
$$

速度势不闭合，对应非零环量，是理想流体产生升力的核心数学机制。

# 模块 3：环量定理与解析性严格证明
## 3 环量定理与解析性严格证明

### 3.1 环量与通量的复积分定义
对任意闭合曲线 \( C \)，定义：

- 环量（切向积分）：\( \Gamma_C = \oint_C (u dx + v dy) \)
- 通量（法向积分）：\( Q_C = \oint_C (u dy - v dx) \)

统一为复积分：
$$
\oint_C V(z) dz = \oint_C (u - iv) d(x+iy) = \Gamma_C + i Q_C
$$
其中 \( V(z) = \frac{d\Phi}{dz} \) 为复速度。

### 3.2 单连通区域解析性充要条件
定理：若 \( D \) 为单连通区域，\( \Phi(z) \) 在 \( D \) 内解析 \( \iff \) 对 \( D \) 内任意闭曲线 \( C \)，\( \Gamma_C=0 \) 且 \( Q_C=0 \)。

#### 证明
- 必要性：\( \Phi(z) \) 解析 \( \Rightarrow V(z) \) 解析，由柯西积分定理：
$$
\oint_C V(z) dz = 0 \Rightarrow \Gamma_C=0,\; Q_C=0
$$
- 充分性：任意闭曲线 \( \Gamma_C=0,\; Q_C=0 \Rightarrow \) 流场无旋无源，存在共轭调和的 \( \varphi \) 与 \( \psi \)，满足 C-R 条件，故 \( \Phi(z) \) 解析。

### 3.3 多连通区域反例与物理说明
多连通区域（\( r>a \)，含原点奇点）定理不成立。

反例：带环量圆柱绕流，取 \( C:|z|=a \)：
$$
\oint_C V(z) dz = -i\Gamma \Rightarrow \Gamma_C = -\Gamma \neq 0
$$
但 \( \Phi(z) \) 在 \( r>a \) 内依然解析。

物理意义：多连通区域允许存在涡奇点，非零环量是产生升力的必要条件。

# 模块 4：极坐标速度场推导
## 4 极坐标速度场推导

### 4.1 速度势与流函数偏导求解
极坐标下速度分量为：
$$
v_r = \frac{\partial \varphi}{\partial r}, \quad v_\theta = \frac{1}{r}\frac{\partial \varphi}{\partial \theta}
$$

代入速度势 $\varphi$，化简得：
$$
v_r = U\left(1-\frac{a^2}{r^2}\right)\cos\theta
$$
$$
v_\theta = -U\left(1+\frac{a^2}{r^2}\right)\sin\theta - \frac{\Gamma}{2\pi r}
$$

### 4.2 圆柱表面边界条件验证
圆柱表面 $r=a$，径向速度：
$$
\left.v_r\right|_{r=a} = 0
$$

严格满足无穿透边界条件，验证复势构造正确。

### 4.3 环量对速度场的影响
周向速度 $v_\theta$ 新增环量项 $-\frac{\Gamma}{2\pi r}$，打破上下对称，使圆柱上下表面速度差增大，为压力差与升力产生提供基础。

# 模块 5：环量守恒严格推导与验证
## 5 环量守恒严格推导与验证

### 5.1 复速度计算
对复势求导得复速度：
$$
V(z) = \frac{d\Phi}{dz} = U\left(1-\frac{a^2}{z^2}\right) + \frac{i\Gamma}{2\pi z}
$$

### 5.2 闭合曲线积分证明
取圆柱表面 \( C: z=ae^{i\theta} \)，\( dz=iae^{i\theta}d\theta \)，积分：
$$
\oint_C V(z)dz
= \oint_C U\left(1-\frac{a^2}{z^2}\right)dz
+ \oint_C \frac{i\Gamma}{2\pi z}dz
$$

- 第一项：解析函数积分，结果为 \( 0 \)；
- 第二项：由柯西积分公式，结果为 \( -i\Gamma \)。

最终：
$$
\oint_C V(z)dz = -i\Gamma
\Rightarrow \Gamma_C = -\Gamma
$$

### 5.3 环量守恒结论
- 环量积分与路径半径无关，仅由 \( \Gamma \) 决定；
- 环量不随时间、路径变化，满足涡量守恒定律；
- 为库塔‑儒可夫斯基升力定理 \( L=\rho U\Gamma \) 提供严格理论支撑。


# 模块 6：符号推导验证（代码块）
## 6.1 符号推导验证（SymPy）
```python
import sympy as sp

r, theta, U, a, Gamma = sp.symbols('r theta U a Gamma', real=True, positive=True)

# 速度势
phi = U * (r + a**2 / r) * sp.cos(theta) - (Gamma / (2 * sp.pi)) * theta

# 径向速度
v_r = sp.diff(phi, r)

# 周向速度
v_theta = (1 / r) * sp.diff(phi, theta)

print("=" * 60)
print("带环量圆柱绕流极坐标速度分量符号验证结果")
print("=" * 60)

print("\n【1】径向速度 v_r = ")
sp.pretty_print(sp.simplify(v_r))

print("\n【2】周向速度 v_theta = ")
sp.pretty_print(sp.simplify(v_theta))

print("\n【3】圆柱表面(r=a)径向速度验证：")
v_r_surface = sp.simplify(v_r.subs(r, a))
sp.pretty_print(v_r_surface)
```

# 模块 7：环量守恒数值验证（代码块）
## 6.2 环量守恒数值验证
```python
import numpy as np

def verify_circulation_conservation(U, a, Gamma, N=20000):
    theta = np.linspace(0, 2 * np.pi, N, endpoint=False)
    z = a * np.exp(1j * theta)
    dtheta = 2 * np.pi / N
    dz = 1j * a * np.exp(1j * theta) * dtheta
    
    V = U * (1 - a**2 / z**2) + (1j * Gamma) / (2 * np.pi * z)
    integral = np.sum(V * dz)
    calc_gamma = np.real(integral)
    error = np.abs(calc_gamma + Gamma)
    
    return calc_gamma, error

# 标准参数
U = 1.0
a = 1.0
Gamma = 2.0

calc_gamma, error = verify_circulation_conservation(U, a, Gamma, N=20000)

print("\n环量守恒数值验证结果(高精度)")
print(f"理论环量 Γ = {Gamma}")
print(f"数值计算环量 = {calc_gamma:.10f}")
print(f"绝对误差 = {error:.2e}")
print(f"误差是否<1e-10: {error < 1e-10}")
print("\n验证结论: 环量严格守恒!")
```

# 模块 8：速度势周期跃变验证（代码块）
## 6.3 速度势周期跃变验证
```python
import numpy as np

def verify_phi_jump(U, a, Gamma, r0=2.0):
    def velocity_potential(r, theta):
        return U * (r + a**2 / r) * np.cos(theta) - (Gamma / (2 * np.pi)) * theta
    
    theta1 = 0.0
    theta2 = 2 * np.pi
    phi1 = velocity_potential(r0, theta1)
    phi2 = velocity_potential(r0, theta2)
    jump = phi2 - phi1
    theoretical_jump = -Gamma
    error = np.abs(jump - theoretical_jump)
    
    return jump, theoretical_jump, error

U = 1.0
a = 1.0
Gamma = 4.0

jump, theo_jump, error = verify_phi_jump(U, a, Gamma)

print("\n速度势周期跃变验证结果")
print(f"设定环量 Γ = {Gamma}")
print(f"理论周期跃变 = {theo_jump}")
print(f"数值计算周期跃变 = {jump:.10f}")
print(f"绝对误差 = {error:.2e}")
print(f"验证通过: {error < 1e-10}")
print("\n结论: 速度势周期跃变严格等于 -Γ")
```
